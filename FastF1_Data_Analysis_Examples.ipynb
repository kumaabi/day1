```
# -*- coding: utf-8 -*-
"""
FastF1 Data Analysis Examples

This script demonstrates various data analysis tasks that can be performed
using the FastF1 Python package, relying solely on data available through it.

Requirements:
- Python 3.8+
- FastF1: pip install fastf1
- Matplotlib: pip install matplotlib
- Pandas: pip install pandas

Setup:
1. Create a directory for caching data (e.g., 'fastf1_cache').
2. Update the `cache_path` variable below to point to your cache directory.
3. Modify the `year`, `event_name`, `session_type`, `driver1`, and `driver2`
   variables to select the specific event and drivers you want to analyze.
"""
!pip install fastf1
import fastf1
import fastf1.plotting
import matplotlib.pyplot as plt
import pandas as pd
import os # Import os to check for cache directory

# --- Configuration ---
cache_path = 'fastf1_cache'  # IMPORTANT: Update this path to your cache directory
year = 2023
event_name = 'Monaco Grand Prix' # Can be event name or round number (e.g., 5 for Baku)
session_type = 'R' # 'FP1', 'FP2', 'FP3', 'Q', 'S' (Sprint), 'R' (Race)
driver1 = 'VER' # Driver abbreviation
driver2 = 'ALO' # Driver abbreviation

# --- Setup ---

# Create cache directory if it doesn't exist
if not os.path.exists(cache_path):
    try:
        os.makedirs(cache_path)
        print(f"Created cache directory: {cache_path}")
    except OSError as e:
        print(f"Error creating cache directory {cache_path}: {e}")
        print("Please create the directory manually or specify a different path.")
        exit() # Exit if cache directory cannot be created

# Enable the cache
try:
    fastf1.Cache.enable_cache(cache_path)
    print(f"Cache enabled at: {fastf1.Cache.get_cache_path()}")
except Exception as e:
    print(f"Cache setup failed: {e}. Ensure the path exists and is writable.")
    print("Proceeding without cache. Data will be downloaded each time.")

# Enable Matplotlib patches for plotting time deltas and set style
fastf1.plotting.setup_mpl(mpl_timedelta_support=True, color_scheme='fastf1', misc_mpl_mods=True)

# --- Load Session Data ---
print(f"Loading session: {year} {event_name} - {session_type}")
try:
    # Use get_session to load the session data
    session = fastf1.get_session(year, event_name, session_type)
    # Load lap data, telemetry, weather, etc. Set telemetry=True for telemetry-based analyses
    # Note: Loading telemetry can take time and memory. Set to False if not needed.
    session.load(laps=True, telemetry=True, weather=True, messages=True)
    print(f"Loaded session: {session.event['EventName']} {session.name}")
    # Access laps data
    laps = session.laps
except Exception as e:
    print(f"Error loading session {year} {event_name} {session_type}: {e}")
    print("Please check the year, event name/round number, session type, and your internet connection.")
    exit() # Exit if session fails to load

# --- Analysis Examples ---

# 1. Load Race Schedule for a Season
print("\n--- 1. Race Schedule ---")
try:
    schedule = fastf1.get_event_schedule(year)
    # Display relevant columns for the first 5 events
    print(schedule[['RoundNumber', 'Country', 'Location', 'OfficialEventName', 'EventDate', 'Session5Date']].head())
except Exception as e:
    print(f"Error loading schedule for {year}: {e}")

# 2. Get Race Results
print("\n--- 2. Race Results ---")
if hasattr(session, 'results') and not session.results.empty:
    # Select specific columns of interest from the results DataFrame
    results = session.results[['DriverNumber', 'Abbreviation', 'TeamName', 'Position', 'GridPosition', 'Points', 'Status', 'Time']]
    print(results.head())
else:
    print("Session results data not available or empty.")

# 3. Get Fastest Lap of the Race
print("\n--- 3. Fastest Lap ---")
if not laps.empty:
    fastest_lap = laps.pick_fastest()
    if fastest_lap is not None and not fastest_lap.empty:
        print(f"Fastest Lap: {fastest_lap['Driver']} ({fastest_lap['Team']}) - Lap {fastest_lap['LapNumber']} - Time: {fastest_lap['LapTime']}")
        # Display details about the fastest lap
        print(fastest_lap[['Driver', 'Team', 'LapNumber', 'LapTime', 'Stint', 'TyreLife', 'Compound', 'Sector1Time', 'Sector2Time', 'Sector3Time']])
    else:
        print("Could not determine the fastest lap.")
else:
    print("Lap data is empty, cannot determine fastest lap.")

# 4. Compare Lap Times for Two Drivers
print(f"\n--- 4. Lap Time Comparison ({driver1} vs {driver2}) ---")
if not laps.empty:
    laps_driver1 = laps.pick_driver(driver1)
    laps_driver2 = laps.pick_driver(driver2)

    if not laps_driver1.empty and not laps_driver2.empty:
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.plot(laps_driver1['LapNumber'], laps_driver1['LapTime'], label=driver1, color=fastf1.plotting.driver_color(driver1))
        ax.plot(laps_driver2['LapNumber'], laps_driver2['LapTime'], label=driver2, color=fastf1.plotting.driver_color(driver2))
        ax.set_xlabel("Lap Number")
        ax.set_ylabel("Lap Time (MM:SS.ms)")
        ax.set_title(f"{driver1} vs {driver2} Lap Times - {session.event['EventName']} {year}")
        ax.legend()
        plt.grid(True, axis='y')
        plt.tight_layout()
        plt.show()
    else:
        print(f"Could not find lap data for one or both drivers: {driver1}, {driver2}")
else:
    print("Lap data is empty, cannot compare lap times.")

# 5. Plot Telemetry for a Driver's Fastest Lap (Speed)
print(f"\n--- 5. Fastest Lap Telemetry (Speed) for {driver1} ---")
if not laps.empty:
    fastest_lap_driver1 = laps.pick_driver(driver1).pick_fastest()
    if fastest_lap_driver1 is not None and not fastest_lap_driver1.empty:
        # Get car data for the lap and add distance column
        telemetry = fastest_lap_driver1.get_car_data().add_distance()
        if not telemetry.empty:
            fig, ax = plt.subplots(figsize=(12, 5))
            ax.plot(telemetry['Distance'], telemetry['Speed'], label='Speed', color=fastf1.plotting.driver_color(driver1))
            ax.set_xlabel('Distance (m)')
            ax.set_ylabel('Speed (Km/h)')
            ax.set_title(f"{driver1} Fastest Lap ({fastest_lap_driver1['LapTime']}) Telemetry - {session.event['EventName']} {year}")
            ax.legend()
            plt.grid(True)
            plt.tight_layout()
            plt.show()
        else:
            print(f"No telemetry data found for {driver1}'s fastest lap.")
    else:
        print(f"Could not find fastest lap for {driver1}.")
else:
    print("Lap data is empty, cannot plot telemetry.")


# 6. Compare Speed Traces for Two Drivers on their Fastest Laps
print(f"\n--- 6. Speed Trace Comparison ({driver1} vs {driver2}) ---")
if not laps.empty:
    fastest_d1 = laps.pick_driver(driver1).pick_fastest()
    fastest_d2 = laps.pick_driver(driver2).pick_fastest()

    if fastest_d1 is not None and not fastest_d1.empty and fastest_d2 is not None and not fastest_d2.empty:
        telemetry_d1 = fastest_d1.get_car_data().add_distance()
        telemetry_d2 = fastest_d2.get_car_data().add_distance()

        if not telemetry_d1.empty and not telemetry_d2.empty:
            fig, ax = plt.subplots(figsize=(12, 6))
            ax.plot(telemetry_d1['Distance'], telemetry_d1['Speed'], label=f"{driver1} ({fastest_d1['LapTime']})", color=fastf1.plotting.driver_color(driver1))
            ax.plot(telemetry_d2['Distance'], telemetry_d2['Speed'], label=f"{driver2} ({fastest_d2['LapTime']})", color=fastf1.plotting.driver_color(driver2))
            ax.set_xlabel('Distance (m)')
            ax.set_ylabel('Speed (Km/h)')
            ax.set_title(f"Fastest Lap Speed Comparison - {session.event['EventName']} {year}")
            ax.legend()
            plt.grid(True)
            plt.tight_layout()
            plt.show()
        else:
            print(f"Could not retrieve telemetry for one or both drivers' fastest laps.")
    else:
        print(f"Could not find fastest lap for one or both drivers: {driver1}, {driver2}")
else:
    print("Lap data is empty, cannot compare speed traces.")

# 7. Analyze Tyre Strategy (Compounds Used and Stint Lengths)
print(f"\n--- 7. Tyre Strategy Analysis for {driver1} ---")
if not laps.empty:
    laps_driver1 = laps.pick_driver(driver1)
    if not laps_driver1.empty:
        driver_stints = laps_driver1[['Stint', 'Compound', 'LapNumber', 'TyreLife']].dropna(subset=['Stint'])
        stints = driver_stints.groupby('Stint')

        print(f"Tyre Stints for {driver1}:")
        for stint_num, stint_data in stints:
            compound = stint_data['Compound'].iloc[0]
            start_lap = stint_data['LapNumber'].min()
            end_lap = stint_data['LapNumber'].max()
            length = (end_lap - start_lap) + 1
            start_tyre_life = stint_data['TyreLife'].min()
            end_tyre_life = stint_data['TyreLife'].max()
            print(f"  Stint {int(stint_num)}: Laps {start_lap}-{end_lap} ({length} laps) on {compound} (Tyre Life {int(start_tyre_life)}-{int(end_tyre_life)})")
    else:
        print(f"No lap data found for {driver1}.")
else:
    print("Lap data is empty, cannot analyze tyre strategy.")

# 8. Plot Gear Shifts on Fastest Lap
print(f"\n--- 8. Gear Shift Analysis for {driver1} ---")
if not laps.empty:
    fastest_lap_driver1 = laps.pick_driver(driver1).pick_fastest()
    if fastest_lap_driver1 is not None and not fastest_lap_driver1.empty:
        telemetry_gear = fastest_lap_driver1.get_car_data().add_distance()
        if not telemetry_gear.empty and 'nGear' in telemetry_gear.columns:
            fig, ax = plt.subplots(figsize=(12, 4))
            ax.plot(telemetry_gear['Distance'], telemetry_gear['nGear'], label='Gear', color='green', marker='.', linestyle='')
            ax.set_xlabel('Distance (m)')
            ax.set_ylabel('Gear')
            ax.set_yticks(sorted(telemetry_gear['nGear'].unique())) # Set y-ticks to actual gears used
            ax.set_title(f"{driver1} Fastest Lap ({fastest_lap_driver1['LapTime']}) Gear Shifts - {session.event['EventName']} {year}")
            plt.grid(True)
            plt.tight_layout()
            plt.show()
        else:
            print(f"No telemetry or gear data found for {driver1}'s fastest lap.")
    else:
        print(f"Could not find fastest lap for {driver1}.")
else:
    print("Lap data is empty, cannot plot gear shifts.")

# --- Analysis Examples ---
# ... (rest of the code is the same)


# 10. Compare Fastest Sector Times of Top Drivers
print("\n--- 10. Fastest Sector Times Comparison ---")
if not laps.empty:
    # Get fastest overall lap first
    fastest_overall = laps.pick_fastest()
    if fastest_overall is not None and not fastest_overall.empty:
        print(f"Overall Fastest Lap: {fastest_overall['Driver']} {fastest_overall['LapTime']}")
        print(f"  Sector 1: {fastest_overall['Sector1Time']}")
        print(f"  Sector 2: {fastest_overall['Sector2Time']}")
        print(f"  Sector 3: {fastest_overall['Sector3Time']}")

        # Find fastest sector times across all laps/drivers
        fastest_s1_lap = laps.sort_values(by='Sector1Time').iloc[0]
        fastest_s2_lap = laps.sort_values(by='Sector2Time').iloc[0]
        fastest_s3_lap = laps.sort_values(by='Sector3Time').iloc[0]

        print("\nFastest Individual Sectors:")
        print(f"  Sector 1: {fastest_s1_lap['Sector1Time']} by {fastest_s1_lap['Driver']} on Lap {fastest_s1_lap['LapNumber']}")
        print(f"  Sector 2: {fastest_s2_lap['Sector2Time']} by {fastest_s2_lap['Driver']} on Lap {fastest_s2_lap['LapNumber']}")
        print(f"  Sector 3: {fastest_s3_lap['Sector3Time']} by {fastest_s3_lap['Driver']} on Lap {fastest_s3_lap['LapNumber']}")
    else:
        print("Could not determine the overall fastest lap.")
else:
    print("Lap data is empty, cannot analyze sector times.")


# 11. Driver Positions Gained/Lost in the Race
print("\n--- 11. Positions Gained/Lost ---")
if hasattr(session, 'results') and not session.results.empty and 'GridPosition' in session.results.columns and 'Position' in session.results.columns:
    results_pos = session.results.copy() # Work on a copy
    # Convert positions to numeric, coercing errors for non-finishers etc.
    results_pos['GridPositionNum'] = pd.to_numeric(results_pos['GridPosition'], errors='coerce')
    results_pos['PositionNum'] = pd.to_numeric(results_pos['Position'], errors='coerce')

    # Handle pit lane starts (GridPosition 0) - requires qualifying data for accurate starting position
    # Simple approach: treat 0 as last place for calculation if quali data isn't loaded
    num_starters = results_pos['GridPositionNum'].nunique() # Estimate number of starters
    results_pos['GridPositionCalc'] = results_pos['GridPositionNum'].replace(0, num_starters + 1) # Approximate pit lane start position

    # Calculate change only for drivers with valid start and end positions
    results_pos['PositionChange'] = results_pos['GridPositionCalc'] - results_pos['PositionNum']
    results_pos.dropna(subset=['PositionChange'], inplace=True) # Remove drivers where calculation failed

    print(results_pos[['Abbreviation', 'GridPosition', 'Position', 'PositionChange']].sort_values('PositionChange', ascending=False))
else:
    print("Results data or required columns (GridPosition, Position) not available.")

# 12. Track Status Changes (Safety Cars, Flags)
print("\n--- 12. Track Status ---")
# Track status data might need specific loading or isn't always available
if hasattr(session, 'track_status') and session.track_status is not None and not session.track_status.empty:
    print(session.track_status)
    # Example: Plot safety car periods
    sc_periods = session.track_status[session.track_status['Status'] == '4'] # Status 4 often indicates Safety Car
    if not sc_periods.empty:
        print("\nSafety Car Periods:")
        print(sc_periods)
        # Plotting SC periods would require visualizing time intervals on a timeline
else:
    print("Track status data not available or empty for this session.")


# 13. Race Control Messages
print("\n--- 13. Race Control Messages ---")
if hasattr(session, 'messages') and not session.messages.empty:
    rcm = session.messages
    print(f"Total Race Control Messages: {len(rcm)}")
    print(rcm.head()) # Show first few messages

    # Example: Filter for specific keywords (e.g., investigations, penalties)
    keywords = ['INVESTIGATED', 'PENALTY', 'UNDER INVESTIGATION', 'LAP DELETED']
    filtered_msgs = rcm[rcm['Message'].str.contains('|'.join(keywords), case=False, na=False)]
    if not filtered_msgs.empty:
        print("\nFiltered Messages (Investigations/Penalties):")
        print(filtered_msgs[['Time', 'Category', 'Message']])
    else:
        print("\nNo messages found matching keywords:", keywords)
else:
    print("Race control messages not available or empty for this session.")

# 14. Weather Data Overview and Plot
print("\n--- 14. Weather Data ---")
if hasattr(session, 'weather_data') and not session.weather_data.empty:
    weather_data = session.weather_data
    print("Weather Data Sample:")
    print(weather_data.head())

    # Plot Air Temperature and Track Temperature over time
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.plot(weather_data['Time'], weather_data['AirTemp'], label='Air Temp (°C)', color='red')
    ax.plot(weather_data['Time'], weather_data['TrackTemp'], label='Track Temp (°C)', color='blue')
    ax.set_xlabel("Session Time")
    ax.set_ylabel("Temperature (°C)")
    ax.set_title(f"Weather Conditions - {session.event['EventName']} {year}")
    ax.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("Weather data not available or empty for this session.")


# 15. Plot Throttle and Brake Application on Fastest Lap
print(f"\n--- 15. Throttle/Brake Trace for {driver1} ---")
if not laps.empty:
    fastest_lap_driver1 = laps.pick_driver(driver1).pick_fastest()
    if fastest_lap_driver1 is not None and not fastest_lap_driver1.empty:
        telemetry_inputs = fastest_lap_driver1.get_car_data().add_distance()
        if not telemetry_inputs.empty and 'Throttle' in telemetry_inputs.columns and 'Brake' in telemetry_inputs.columns:
            fig, ax = plt.subplots(2, 1, sharex=True, figsize=(12, 7))

            # Throttle Plot
            ax[0].plot(telemetry_inputs['Distance'], telemetry_inputs['Throttle'], label='Throttle', color='lime')
            ax[0].set_ylabel('Throttle (%)')
            ax[0].set_title(f"{driver1} Fastest Lap ({fastest_lap_driver1['LapTime']}) Inputs - {session.event['EventName']} {year}")
            ax[0].legend()
            ax[0].grid(True)

            # Brake Plot (Brake data is often boolean 0/1, sometimes pressure 0-100)
            ax[1].plot(telemetry_inputs['Distance'], telemetry_inputs['Brake'], label='Brake', color='red')
            ax[1].set_xlabel('Distance (m)')
            ax[1].set_ylabel('Brake Applied') # Label might need adjustment based on data (e.g., 'Brake Pressure')
            ax[1].legend()
            ax[1].grid(True)

            plt.tight_layout()
            plt.show()
        else:
            print(f"No telemetry or Throttle/Brake data found for {driver1}'s fastest lap.")
    else:
        print(f"Could not find fastest lap for {driver1}.")
else:
    print("Lap data is empty, cannot plot inputs.")

# 16. Lap Time Distribution (Histogram)
print(f"\n--- 16. Lap Time Distribution for {driver1} ---")
if not laps.empty:
    # Pick accurate laps (exclude in/out laps) and drop laps with no time
    laps_driver1_valid = laps.pick_driver(driver1).pick_accurate().dropna(subset=['LapTime'])
    if not laps_driver1_valid.empty:
        # Convert LapTime timedelta to seconds for plotting
        laps_driver1_valid['LapTimeSeconds'] = laps_driver1_valid['LapTime'].dt.total_seconds()

        fig, ax = plt.subplots(figsize=(10, 6))
        ax.hist(laps_driver1_valid['LapTimeSeconds'], bins=20, alpha=0.7, color=fastf1.plotting.driver_color(driver1), edgecolor='black')
        ax.set_xlabel("Lap Time (seconds)")
        ax.set_ylabel("Frequency (Number of Laps)")
        ax.set_title(f"{driver1} Lap Time Distribution (Valid Laps) - {session.event['EventName']} {year}")
        plt.grid(True, axis='y')
        plt.tight_layout()
        plt.show()
    else:
        print(f"No valid laps with time data found for {driver1}.")
else:
    print("Lap data is empty, cannot plot lap time distribution.")


# 17. Plotting the Track Map using Fastest Lap Position Data
print("\n--- 17. Track Map ---")
if not laps.empty:
    # Use position data from the fastest lap to draw the track
    fastest_lap_pos = laps.pick_fastest()
    if fastest_lap_pos is not None and not fastest_lap_pos.empty:
        lap_telemetry = fastest_lap_pos.get_pos_data() # Gets X, Y, Z coordinates
        if not lap_telemetry.empty and 'X' in lap_telemetry.columns and 'Y' in lap_telemetry.columns:
            fig, ax = plt.subplots(figsize=(8, 8))
            # Plot the X and Y coordinates
            ax.plot(lap_telemetry['X'], lap_telemetry['Y'], color='grey', linewidth=2, label='Track Layout')
            # Ensure equal aspect ratio
            ax.set_aspect('equal', adjustable='box')
            # Add labels and title
            ax.set_xlabel("X Coordinate (m)")
            ax.set_ylabel("Y Coordinate (m)")
            ax.set_title(f"{session.event['EventName']} Track Layout ({year})")
            # Hide axis ticks for a cleaner look
            plt.xticks([])
            plt.yticks([])
            plt.tight_layout()
            plt.show()
        else:
            print("Position data (X, Y coordinates) not found for the fastest lap.")
    else:
        print("Could not determine the fastest lap to plot track map.")
else:
    print("Lap data is empty, cannot plot track map.")

# 18. Average Speed per Lap Comparison
print(f"\n--- 18. Average Speed per Lap ({driver1} vs {driver2}) ---")
if not laps.empty:
    avg_speeds_d1 = []
    avg_speeds_d2 = []

    # Calculate for Driver 1
    valid_laps_d1 = laps.pick_driver(driver1).pick_accurate()
    if not valid_laps_d1.empty:
        for lap_info in valid_laps_d1.iterlaps():
            lap = lap_info[1] # Get the Lap object
            try:
                telemetry = lap.get_car_data()
                if not telemetry.empty and 'Speed' in telemetry.columns:
                    avg_speed = telemetry['Speed'].mean()
                    avg_speeds_d1.append({'LapNumber': lap['LapNumber'], 'AvgSpeed': avg_speed})
            except Exception as e:
                # print(f"Warning: Could not get telemetry for {driver1} Lap {lap['LapNumber']}: {e}")
                pass # Silently ignore laps where telemetry fails
        avg_speed_df_d1 = pd.DataFrame(avg_speeds_d1)
    else:
        print(f"No valid laps found for {driver1}")
        avg_speed_df_d1 = pd.DataFrame() # Empty dataframe

    # Calculate for Driver 2
    valid_laps_d2 = laps.pick_driver(driver2).pick_accurate()
    if not valid_laps_d2.empty:
        for lap_info in valid_laps_d2.iterlaps():
            lap = lap_info[1]
            try:
                telemetry = lap.get_car_data()
                if not telemetry.empty and 'Speed' in telemetry.columns:
                    avg_speed = telemetry['Speed'].mean()
                    avg_speeds_d2.append({'LapNumber': lap['LapNumber'], 'AvgSpeed': avg_speed})
            except Exception as e:
                # print(f"Warning: Could not get telemetry for {driver2} Lap {lap['LapNumber']}: {e}")
                pass
        avg_speed_df_d2 = pd.DataFrame(avg_speeds_d2)
    else:
         print(f"No valid laps found for {driver2}")
         avg_speed_df_d2 = pd.DataFrame() # Empty dataframe

    # Plotting
    if not avg_speed_df_d1.empty or not avg_speed_df_d2.empty:
        fig, ax = plt.subplots(figsize=(12, 6))
        if not avg_speed_df_d1.empty:
            ax.plot(avg_speed_df_d1['LapNumber'], avg_speed_df_d1['AvgSpeed'], label=f'{driver1} Avg Speed', marker='.', linestyle='-', color=fastf1.plotting.driver_color(driver1))
        if not avg_speed_df_d2.empty:
            ax.plot(avg_speed_df_d2['LapNumber'], avg_speed_df_d2['AvgSpeed'], label=f'{driver2} Avg Speed', marker='.', linestyle='-', color=fastf1.plotting.driver_color(driver2))

        ax.set_xlabel("Lap Number")
        ax.set_ylabel("Average Speed (Km/h)")
        ax.set_title(f"Average Speed per Lap Comparison - {session.event['EventName']} {year}")
        ax.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print("Could not calculate average speeds for plotting.")
else:
    print("Lap data is empty, cannot calculate average speeds.")


print("\n--- Analysis Complete ---")
```
